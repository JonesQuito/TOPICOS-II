# Importar o módulo do postgres

```python
import psycopg2
```

# Criar uma conexão e fazer uma consulta
 O código a seguir assume que existe um banco de dados do postgres em host=`localhost`, com nome=`topicos_ii`, user=`postgres` e password=`postgres`. Assume também que exite uma tabela com seguinte estrutura: `create table classe(id serial primary key, nome varchar(30), descricao varchar(200))`.<br>
A consulta a seguir finaliza com um for para percorrer todos os elementos retornados no resultset. Após o laço a conexão é fechada com `con.close()`</br>
 

```python
# Criar uma conexão e criar a tabela 'classe'

con = psycopg2.connect(host='localhost', database='topicos_ii', user='postgres', password='postgres')
cursor = con.cursor()
sql = 'select * from classe'
cursor.execute(sql)

recset = cursor.fetchall()
for rec in recset:
    print(rec)

con.close()
```

```python
# Atualizando a descrição da classe 'saúde'
con = psycopg2.connect(host='localhost', database='topicos_ii', user='postgres', password='postgres')
cursor = con.cursor()
sql = "update classe set descricao = 'Textos sobre saúde' where nome='saude'"
cursor.execute(sql)
con.commit()
con.close()
```
